```python
import matplotlib.pyplot as plt
from app.qkd.bb84 import run_bb84
from app.qkd.eve import run_bb84_with_eve
from app.qkd.utils import print_key_summary


# Simulate secure communication
print("\n--- Scenario 1: BB84 Key Exchange (No Eavesdropping) ---")
alice_key, bob_key, indices = run_bb84(200)
print_key_summary(alice_key, bob_key)

# Simulate attack
print("\n--- Scenario 2: BB84 Key Exchange (With Eve Eavesdropping) ---")
a_key_e, b_key_e, idx_e, err_rate = run_bb84_with_eve(200)
print(f"Resulting Key Error Rate due to Eve's Interference: {err_rate:.2%}")
print_key_summary(a_key_e, b_key_e)

# Visualization
plt.bar(["No Eve", "With Eve"], [1.0, 1.0 - err_rate], color=["green", "red"])
plt.ylabel("Key Agreement Ratio")
plt.title("Impact of Eavesdropping on BB84 Protocol")
plt.show()
```
